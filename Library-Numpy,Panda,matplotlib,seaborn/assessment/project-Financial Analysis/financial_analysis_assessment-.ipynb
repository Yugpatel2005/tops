# Insightful Analysis Questions

## 1. What percentage of the dataset shows negative GDP growth?

```python
((df['GDP Growth (%)'] < 0).mean())*100
```

## 2. Does high inflation correspond to higher interest rates?

```python
# Correlation matrix
corr = df[['Inflation Rate (%)','Interest Rate (%)']].corr()

# Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title("Correlation between Inflation Rate and Interest Rate")
plt.show()
```

## 3. Is there a relationship between unemployment and consumer spending?

```python
df['Unemployment Rate (%)'].corr(df['Consumer Spending (Billion USD)'])
```

## 4. Do higher corporate profits align with higher consumer confidence?

```python
df['Corporate Profits (Billion USD)'].corr(df['Consumer Confidence Index'])
```

## 5. What’s the trend of crude oil prices over time?

```python

df['Date'] = pd.to_datetime(df['Date'])

# Year column
df['Year'] = df['Date'].dt.year

yearly = df.groupby('Year')['Crude Oil Price (USD per Barrel)'].mean()

plt.plot(yearly.index, yearly.values, marker='o')
plt.xlabel("Year")
plt.ylabel("Crude Oil Price (USD per Barrel)")
plt.title("Year-wise Crude Oil Price Trend")
plt.show()
```
