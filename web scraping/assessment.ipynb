```python
import requests
from bs4 import BeautifulSoup
```

```python
# Step 1: URL of website
url = "https://books.toscrape.com/"
```

```python
# Step 2: Send request to website
response = requests.get(url)
```

```python
response.status_code
```

```python
# Step 3: Create BeautifulSoup object
soup = BeautifulSoup(response.text, "html.parser")
soup
```

```python

# Step 4: Find all books category
books_category = soup.find("div",class_="side_categories")

title = books_category.find_all("a")
for i in title:
    print(i.get_text(strip=True))
    
```

```python
# Step 5: Find all books blocks
books_article = soup.find_all("article", class_="product_pod")
```

```python
# Step 6: Extract data
i=1
for q in books_article:
  
  name = q.find("h3").get_text(strip=True)
  price = q.find("p",class_='price_color').get_text(strip=True)
  stock = q.find("p",class_='instock availability').get_text(strip=True)
  print(f"{i} book name:", name)
  i+=1
  print("price:", price)
  print("stock", stock)
  print("------------")
    
```

```python
## PAGINATION

import csv

base_url = "https://books.toscrape.com/catalogue/page-{}.html"

all_book =[]

for page in range(1,51):
    print(f"Scraping page {page}")
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text,"html.parser")

    books = soup.find_all("article",class_="product_pod")

    for book in books:
        title = book.h3.a["title"]
        price = book.find("p",class_="price_color").text
        availability = book.find("p",class_="instock availability").text.strip()
        rating = book.p["class"][1]
        book_url = "https://books.toscrape.com/catalogue/" + book.h3.a["href"]

        all_book.append([title,price,availability,rating,book_url])
        
```

```python
import csv

with open("1000_books_list1.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file, delimiter=",")
    writer.writerow(["Title", "Price", "Availability", "Rating", "Book URL"])
    writer.writerows(all_book)


print("1000 books successfully scraped and saved to csv")
```
